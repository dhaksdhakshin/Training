#  Task 1: Simple Plot

```python
import matplotlib.pyplot as plt


x_values=[1,2,3]

y_values=[4,5,6]


plt.plot(x_values, y_values, marker='o', linestyle='-', color='b',  label="y = 2x")


plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Simple Plot')

plt.legend
plt.grid(True)
plt.show()
```

#  Task 3: Function Calculator

```python
def calc_area(length, width=2):
    return length * width


result = calc_area(6)

print(result)
```

# Task 4: Lambda Transformation

```python
numbers=[1,2,3,4,5]

even_numbers=list(filter(lambda x: x % 2 ==0, numbers))

cubed_numbers=list(map(lambda x: x ** 3, even_numbers))

print(" Cubed Even Numbers", cubed_numbers)
```

# Bonus: Advanced EDA

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidth=0.5)

plt.title("Correlation Heatmap")
plt.show()
```
